## REGRESION LINEAL


```python
# Grafica de dispersion, salario vs experiencia.
y = salarios["Salario"]
x = salarios["Años de Experiencia"]
#scatter  es un  tipo de grafica de dispersion
#el primer valor es el eje X y el segundo valor es el eje Y para poder generar la grafica

plt.scatter(x,y,color = "red")

#titulo de la grafica

plt.title("Salario vs Años de Experiencia")

#Titulo eje x

plt.xlabel("Años de Experiencia")

#Titulo Eje Y

plt.ylabel("Salario")

plt.show()
```
