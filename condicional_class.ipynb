```python
# 1.Bucle for El bucle for se utiliza para iterar sobre una secuencia,
 #(como una lista, tupla, diccionario, conjunto o cadena de caracteres).
 #Se ejecuta una vez por cada elemento en la secuencia.
```

```python
# Imprimir todos los numero de una secuencia de datos

secuencia=[1,2,3,4,5]

for i in secuencia :

    print (i)
```

```python
# Clasifica los numeros que son pares y los que no son pares de la lista secuencia

for i in secuencia:
    if i % 2 == 0:
     print(f"el numero{i} es par ")
    else:
     print(f"el numero{i} es in par ")
```

```python
# Suma
a = 10
b = 5
print(a + b)  # 15

# Resta
print(a - b)  # 5

# Multiplicación
print(a * b)  # 50

# División
print(a / b)  # 2.0

# División entera
print(a // b)  # 2

# Módulo
print(a % b)  # 0

# Exponenciación
print(a ** b)  # 100000

# Asignación compuesta
a += b
print(a)  # 15

# Redondeo
c = 3.14159
print(round(c, 2))  # 3.14

# Valor absoluto
d = -10
print(abs(d))  # 10

# Máximo y mínimo
e = [1, 2, 3, 4, 5]
print(max(e))  # 5
print(min(e))  # 1

# Potencia
print(pow(2, 3))  # 8
```
